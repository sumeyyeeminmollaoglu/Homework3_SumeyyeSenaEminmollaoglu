```python
from __future__ import print_function, division

%matplotlib inline
import pandas
import numpy as np
import matplotlib
import thinkstats2
import thinkplot
import nsfg
```

```python
resp=pandas.DataFrame(nsfg.ReadFemResp())
```

```python
resp.timesmar
```

```python
timesmar = resp.timesmar.dropna()
```

```python
timesmar_mean, timesmar_std=timesmar.mean(),timesmar.std()
```

```python
timesmar_mean, timesmar_std
```

```python
jintendn=resp.jintendn.dropna()
jintendn_mean, jintendn_std=jintendn.mean(),jintendn.std()
```

```python
jintendn_mean, jintendn_std
```

```python
pdf_timesmar = thinkstats2.NormalPdf(timesmar_mean, timesmar_std)
pdf_timesmar.Density(timesmar_mean + timesmar_std)
```

```python
thinkplot.Pdf(pdf_timesmar, label='normal')
thinkplot.Config(xlabel='x', ylabel='PDF')
```

```python
pdf_jintendn = thinkstats2.NormalPdf(timesmar_mean, timesmar_std)
pdf_jintendn.Density(jintendn_mean + jintendn_std)
```

```python
thinkplot.Pdf(pdf_jintendn, label='normal')
thinkplot.Config(xlabel='x', ylabel='PDF')
```

```python
def Jitter(values, jitter=0.5):
    n = len(values)
    return np.random.normal(0, jitter, n) + values
```

```python
timesmar = Jitter(timesmar)
jintendn = Jitter(jintendn)
```

```python
thinkplot.Scatter(timesmar, jintendn, alpha=0.01, s=2,color='black')
thinkplot.Config(xlabel='timesmar',
                 ylabel='jintendn',
                 legend=False)
```

```python
cleaned = resp.dropna(subset=['timesmar', 'jintendn'])
cleaned
```

```python
bins = np.arange(1,5)
print(bins)
indices = np.digitize(cleaned.timesmar, bins)
print(indices)
groups = cleaned.groupby(indices)
print(groups)
```

```python
for i, group in groups:
    print(i, len(group))
```

```python
mean_timesmar = [group.timesmar.mean() for i, group in groups]
cdfs = [thinkstats2.Cdf(group.jintendn) for i, group in groups]
mean_timesmar
```

```python
CustomColors = ["black","blue","yellow"]
percent = [25,50,75]
for i in range(len(percent)):
    weight_percentiles = [cdf.Percentile(percent[i]) for cdf in cdfs]
    label = '%dth' % percent[i]
    thinkplot.Plot(mean_timesmar, weight_percentiles, label=label,color=CustomColors[i])
    
thinkplot.Config(xlabel='timesmar',
                 ylabel='jintendn',
                 legend=True)
```
